```python
import pandas as pd

fighter_data = pd.read_pickle("fighter_data.pkl")
fights_data = pd.read_pickle("fights_data_cleaning2.pkl")

# merge on fighter 1
combined_fighter_fights_data1 = pd.merge(fights_data, fighter_data, left_on='fighter 1', right_on='Name')

# clean and rename columns
del combined_fighter_fights_data1['Name']
combined_fighter_fights_data1.rename(columns={'Height': 'Height 1', 'Weight': 'Weight 1', 'Stance': 'Stance 1', 'Record': 'Record 1'}, inplace=True)

# merge on fighter 2
combined_fighter_fights_data_final = pd.merge(combined_fighter_fights_data1, fighter_data, left_on='fighter 2', right_on='Name')

# clean and rename columns
del combined_fighter_fights_data_final['Name']
combined_fighter_fights_data_final.rename(columns={'Height': 'Height 2', 'Weight': 'Weight 2', 'Stance': 'Stance 2', 'Record': 'Record 2'}, inplace=True)


list(combined_fighter_fights_data_final.columns)
combined_fighter_fights_data_final
```

```python
combined_fighter_fights_data_final.to_pickle("fights_data_cleaning2.pkl")
```
