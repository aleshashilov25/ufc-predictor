```python
import numpy as np
import pandas as pd
```

```python
#even more cleaning
fights_data = pd.read_pickle("fights_strings_removed.pkl")
fights_data
```

```python
#first we should encode method of wins
set(fights_data["method1"])
```

```python
set(fights_data["method2"])
```

```python
#lets make all decision types and dq as decision, and all ko/tkos as ko for ease
m1 = []

for m in fights_data["method1"]:
    if m == "DQ":
        m1.append("Decision")
    elif m.split(" ")[0] == "Decision":
        m1.append("Decision")
    elif m == "KO/TKO":
        m1.append("KO")
    elif m == "TKO - Doctor's Stoppage":
        m1.append("KO")
    elif m == "Submission":
        m1.append("Sub")
fights_data["method"] = m1

fights_data = fights_data.drop("method1", axis=1)
fights_data = fights_data.drop("method2", axis=1)
```

```python
#lets drop some other redundant columns
#target variable - result: 0 of fighter 1 loses, 1 if fighter 1 wins

fights_data["result"] = fights_data["result 1"]

fights_data = fights_data.drop("result 1", axis=1)
fights_data = fights_data.drop("result 2", axis=1)
```

```python
fights_data["date"] = fights_data["date1"]

fights_data = fights_data.drop("date1", axis=1)
fights_data = fights_data.drop("date2", axis=1)
```

```python
fights_data["time"] = fights_data["time1"]

fights_data = fights_data.drop("time1", axis=1)
fights_data = fights_data.drop("time2", axis=1)
```

```python
fights_data.info()
```

```python
fights_data[fights_data["fighter 2"] == "Dustin Poirier"]
```

```python
#now dealing with records is a little tricky because this dataset is only ufc fights but the record indicates fights
#amongst previous organizations
#for every fighter let's count number of wins and losses, then subtract from the record
#lets get all unique fighters in this dataset
fighters1 = set(fights_data["fighter 1"])
fighters2 = set(fights_data["fighter 2"])
fighters = list(fighters1)
fighters = fighters + list(fighters2)
fighters = set(fighters)
```

```python
#ufc record dict
ufc_record_dict = {}
for f in fighters:
    ufc_record_dict[f] = [0, 0]
    
for f in fighters:
    tempdf1 = fights_data[fights_data["fighter 1"] == f]
    for r in tempdf1["result"]:
        if r == 0:
            ufc_record_dict[f][1] += 1
        else:
            ufc_record_dict[f][0] += 1
    
    
    tempdf2 = fights_data[fights_data["fighter 2"] == f]
    for r in tempdf2["result"]:
        if r == 0:
            ufc_record_dict[f][0] += 1
        else:
            ufc_record_dict[f][1] += 1
```

```python
ufc_record_dict["Jose Aldo"]
```

```python
#now lets subtract this from fighter records to get an initial record
#lets retrive everyone's record and put it in a dict
fighter_data = pd.read_pickle("fighter_data.pkl")
fighter_data
```

```python
mma_record_dict = {}

for f in fighters:
    mma_record_dict[f] = [0, 0]

for f in fighters:
    temp_df = fighter_data[fighter_data["Name"] == f]
    for r in temp_df["Record"]:
        mma_record_dict[f][0] = int(r[0])
        mma_record_dict[f][1] = int(r[1])
mma_record_dict
```

```python
#initial mma record on joining
init_record_dict = {}

for f in fighters:
    init_record_dict[f] = [mma_record_dict[f][0] - ufc_record_dict[f][0],
                           mma_record_dict[f][1] - ufc_record_dict[f][1]]

init_record_dict
```

```python
#we want to track wins and losses so far
#lets reverse the dataframe so it starts from first fight
fights_data2 = fights_data.loc[::-1, :]
fights_data2
```

```python
#wins and losses list (wins\losses prior to incoming fight)
wins1 = []
losses1 = []
wins2 = []
losses2 = []

for i in fights_data2.index:
    f1 = fights_data2.loc[i, :]["fighter 1"]
    f2 = fights_data2.loc[i, :]["fighter 2"]
    r = fights_data2.loc[i, :]["result"]
    wins1.append(init_record_dict[f1][0])
    losses1.append(init_record_dict[f1][1])
    
    wins2.append(init_record_dict[f2][0])
    losses2.append(init_record_dict[f2][1])
   
    
    if r == 0:
        init_record_dict[f1][1] += 1
        init_record_dict[f2][0] += 1
    else:
        init_record_dict[f1][0] += 1
        init_record_dict[f2][1] += 1

       
```

```python
fights_data2["Wins 1"] = wins1
fights_data2["Wins 2"] = wins2
fights_data2["Losses 1"] = losses1
fights_data2["Losses 2"] = losses2
```

```python
fights_data2[fights_data2["fighter 2"] == "Conor McGregor"]
```

```python
#encode win types
methods = fights_data2["method"].astype("category").cat.codes
#0:decision, 1:ko, 2:sub
```

```python
fights_data2["method"] = methods
```

```python
#now that we have this on in our dataframe, we should similarly count type of wins
kos1= []
subs1 = []
decs1 = []

kos2= []
subs2 = []
decs2 = []

lkos1= []
lsubs1 = []
ldecs1 = []

lkos2= []
lsubs2 = []
ldecs2 = []

wintype_dict = {}
for f in fighters:
    wintype_dict[f] = [0, 0, 0]

losstype_dict = {}
for f in fighters:
    losstype_dict[f] = [0, 0, 0]
    
for i in fights_data2.index:
    f1 = fights_data2.loc[i, :]["fighter 1"]
    f2 = fights_data2.loc[i, :]["fighter 2"]
    r = fights_data2.loc[i, :]["result"]
    m = fights_data2.loc[i, :]["method"]
    
    kos1.append(wintype_dict[f1][1])
    subs1.append(wintype_dict[f1][2])
    decs1.append(wintype_dict[f1][0])
   
    kos2.append(wintype_dict[f2][1])
    subs2.append(wintype_dict[f2][2])
    decs2.append(wintype_dict[f2][0])
    
    lkos1.append(losstype_dict[f1][1])
    lsubs1.append(losstype_dict[f1][2])
    ldecs1.append(losstype_dict[f1][0])
   
    lkos2.append(losstype_dict[f2][1])
    lsubs2.append(losstype_dict[f2][2])
    ldecs2.append(losstype_dict[f2][0])

    
    if r == 0:
        wintype_dict[f2][m] += 1
        losstype_dict[f1][m] += 1
    else:
        wintype_dict[f1][m] += 1
        losstype_dict[f2][m] += 1
        
fights_data2["kos 1"] = kos1
fights_data2["subs 1"] = subs1
fights_data2["decs 1"] = decs1

fights_data2["kos 2"] = kos2
fights_data2["subs 2"] = subs2
fights_data2["decs 2"] = decs2

fights_data2["lkos 1"] = lkos1
fights_data2["lsubs 1"] = lsubs1
fights_data2["ldecs 1"] = ldecs1

fights_data2["lkos 2"] = lkos2
fights_data2["lsubs 2"] = lsubs2
fights_data2["ldecs 2"] = ldecs2
```
