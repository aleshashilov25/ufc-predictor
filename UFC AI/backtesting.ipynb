```python
#backtesting is a little tricky, instead of having the stats that took place in a fight, since we wouldnt have them,
#we would instead need to construct a new dataframe that ahs fighters aggregated stats at the time right before they
#were fighting, to see how the model would truly perform
import numpy as np
import pandas as pd
```

```python
#we will use the fully cleaned dataset for this
model_df = pd.read_pickle("fully_cleaned.pkl")
```

```python
model_df.info()
```

```python
#we will use a dictionary that will be updated everytime we come across a fighter
#lets get all unique fighters first
fighters = set(list(model_df["fighter 1"]) + list(model_df["fighter 2"]))
len(fighters)
#for each fighters lets initialize another dicitionary with the needed stats
#note that we already did this for wintypes and wins so far and so on so we will not include that
backtesting_dict = {}
#initializing everything to 0, as we don't know anything about stats before a fighter's initial fight, lets
#assume fighters are expected to be 'average', after the iteration we will set everything that is 0
#to be the average
for f in fighters:
    backtesting_dict[f] = {}

for f in fighters:
    
    backtesting_dict[f]['Sig. str'] = 0
    backtesting_dict[f]['Sig. str. %'] = 0
    
    backtesting_dict[f]['Head'] = 0
    backtesting_dict[f]['Body'] = 0
    backtesting_dict[f]['Leg'] = 0
    backtesting_dict[f]['Distance'] = 0
    backtesting_dict[f]['Clinch'] = 0
    backtesting_dict[f]['Ground'] = 0
    
    backtesting_dict[f]['KD'] = 0
    backtesting_dict[f]['Total str.'] = 0
    
    backtesting_dict[f]['Td'] = 0
    backtesting_dict[f]['Td %'] = 0
    
    backtesting_dict[f]['Sub. att'] = 0
    backtesting_dict[f]['Rev.'] = 0
    backtesting_dict[f]['Ctrl'] = 0
    
    backtesting_dict[f]['Sig. str attempted'] = 0
    backtesting_dict[f]['Sig. str defended'] = 0
    backtesting_dict[f]['Sig. str absorbed'] = 0
    
    backtesting_dict[f]['Td attempted'] = 0
    backtesting_dict[f]['tdd'] = 0
    
    backtesting_dict[f]['time'] = 0
    
    
```

```python
#initialize lists we need
sig_str1 = []
ssp1 = []

h1 = []
b1 = []
l1 = []
d1 = []
c1 = []
g1 = []

kd1 = []
t_str1 = []

td1 = []
tdp1 = []

sub1 = []
rev1 = []
ctrl1 = []

sig_str_attempted1 = []
sig_str_defended1 = []
sig_str_absorbed1 = []

td_attempted1 = []
tdd1 = []

time1 = []
```

```python
sig_str2 = []
ssp2 = []

h2 = []
b2 = []
l2 = []
d2 = []
c2 = []
g2 = []

kd2 = []
t_str2 = []

td2 = []
tdp2 = []

sub2 = []
rev2 = []
ctrl2 = []

sig_str_attempted2 = []
sig_str_defended2 = []
sig_str_absorbed2 = []

td_attempted2 = []
tdd2 = []

time2 = []
```

```python
#now iterate and update
#we will update percentages later, will be a derived va;ue
for i in model_df.index:
    row = model_df.loc[i]
    f1 = row["fighter 1"]
    f2 = row["fighter 2"]
    
    #first we update lists
    #fighter 1
    sig_str1.append(backtesting_dict[f1]['Sig. str'])
    
    h1.append(backtesting_dict[f1]['Head'])
    b1.append(backtesting_dict[f1]['Body'])
    l1.append(backtesting_dict[f1]['Leg'])
    d1.append(backtesting_dict[f1]['Distance'])
    c1.append(backtesting_dict[f1]['Clinch'])
    g1.append(backtesting_dict[f1]['Ground'])
    
    kd1.append(backtesting_dict[f1]["KD"])
    t_str1.append(backtesting_dict[f1]["Total str."])
    
    td1.append(backtesting_dict[f1]["Td"])
    sub1.append(backtesting_dict[f1]["Sub. att"])
    rev1.append(backtesting_dict[f1]["Rev."])
    ctrl1.append(backtesting_dict[f1]["Ctrl"])
    
    sig_str_attempted1.append(backtesting_dict[f1]["Sig. str attempted"])
    sig_str_defended1.append(backtesting_dict[f1]["Sig. str defended"])
    sig_str_absorbed1.append(backtesting_dict[f1]["Sig. str absorbed"])
    
    td_attempted1.append(backtesting_dict[f1]["Td attempted"])
    tdd1.append(backtesting_dict[f1]["tdd"])
    
    time1.append(backtesting_dict[f1]["time"])
    
    
    
    #fighter 2
    sig_str2.append(backtesting_dict[f2]['Sig. str'])
    
    h2.append(backtesting_dict[f2]['Head'])
    b2.append(backtesting_dict[f2]['Body'])
    l2.append(backtesting_dict[f2]['Leg'])
    d2.append(backtesting_dict[f2]['Distance'])
    c2.append(backtesting_dict[f2]['Clinch'])
    g2.append(backtesting_dict[f2]['Ground'])
    
    kd2.append(backtesting_dict[f2]["KD"])
    t_str2.append(backtesting_dict[f2]["Total str."])
    
    td2.append(backtesting_dict[f2]["Td"])
    sub2.append(backtesting_dict[f2]["Sub. att"])
    rev2.append(backtesting_dict[f2]["Rev."])
    ctrl2.append(backtesting_dict[f2]["Ctrl"])
    
    sig_str_attempted2.append(backtesting_dict[f2]["Sig. str attempted"])
    sig_str_defended2.append(backtesting_dict[f2]["Sig. str defended"])
    sig_str_absorbed2.append(backtesting_dict[f2]["Sig. str absorbed"])
    
    td_attempted2.append(backtesting_dict[f2]["Td attempted"])
    tdd2.append(backtesting_dict[f2]["tdd"])
    
    time2.append(backtesting_dict[f2]["time"])
    
    
    #now we update the dictionary
    #fighter 1
    backtesting_dict[f1]['Sig. str'] = backtesting_dict[f1]['Sig. str'] + row["Sig. str 1"]
    
    backtesting_dict[f1]['Head'] = backtesting_dict[f1]['Head'] + row["Head 1"]
    backtesting_dict[f1]['Body'] = backtesting_dict[f1]['Body'] + row["Body 1"]
    backtesting_dict[f1]['Leg'] = backtesting_dict[f1]['Leg'] + row["Leg 1"]
    backtesting_dict[f1]['Distance'] = backtesting_dict[f1]['Distance'] + row["Distance 1"]
    backtesting_dict[f1]['Clinch'] = backtesting_dict[f1]['Clinch'] + row["Clinch 1"]
    backtesting_dict[f1]['Ground'] = backtesting_dict[f1]['Ground'] + row["Ground 1"]
    
    backtesting_dict[f1]['KD'] = backtesting_dict[f1]['KD'] + row["KD 1"]
    backtesting_dict[f1]['Total str.'] = backtesting_dict[f1]['Total str.'] + row["Total str. 1"]
    
    backtesting_dict[f1]['Td'] = backtesting_dict[f1]['Td'] + row["Td 1"]
    
    backtesting_dict[f1]['Sub. att'] = backtesting_dict[f1]['Sub. att'] + row["Sub. att 1"]
    backtesting_dict[f1]['Rev.'] = backtesting_dict[f1]['Rev.'] + row["Rev. 1"]
    backtesting_dict[f1]['Ctrl'] = backtesting_dict[f1]['Ctrl'] + row["Ctrl 1"]
    
    backtesting_dict[f1]['Sig. str attempted'] = backtesting_dict[f1]['Sig. str attempted'] + row['Sig. str attempted1']
    backtesting_dict[f1]['Sig. str defended'] = backtesting_dict[f1]['Sig. str defended'] + row['Sig. str defended1']
    backtesting_dict[f1]['Sig. str absorbed'] = backtesting_dict[f1]['Sig. str absorbed'] + row['Sig. str absorbed1']
    
    backtesting_dict[f1]['Td attempted'] = backtesting_dict[f1]['Td attempted'] + row["Td attempted 1"]
    backtesting_dict[f1]['tdd'] = backtesting_dict[f1]['tdd'] + row['tdd1']
    
    backtesting_dict[f1]['time'] = backtesting_dict[f1]['time'] + row['time']
    
    
    #fighter 2
    backtesting_dict[f2]['Sig. str'] = backtesting_dict[f2]['Sig. str'] + row["Sig. str 2"]
    
    backtesting_dict[f2]['Head'] = backtesting_dict[f2]['Head'] + row["Head 2"]
    backtesting_dict[f2]['Body'] = backtesting_dict[f2]['Body'] + row["Body 2"]
    backtesting_dict[f2]['Leg'] = backtesting_dict[f2]['Leg'] + row["Leg 2"]
    backtesting_dict[f2]['Distance'] = backtesting_dict[f2]['Distance'] + row["Distance 2"]
    backtesting_dict[f2]['Clinch'] = backtesting_dict[f2]['Clinch'] + row["Clinch 2"]
    backtesting_dict[f2]['Ground'] = backtesting_dict[f2]['Ground'] + row["Ground 2"]
    
    backtesting_dict[f2]['KD'] = backtesting_dict[f2]['KD'] + row["KD 2"]
    backtesting_dict[f2]['Total str.'] = backtesting_dict[f2]['Total str.'] + row["Total str. 2"]
    
    backtesting_dict[f2]['Td'] = backtesting_dict[f2]['Td'] + row["Td 2"]
    
    backtesting_dict[f2]['Sub. att'] = backtesting_dict[f2]['Sub. att'] + row["Sub. att 2"]
    backtesting_dict[f2]['Rev.'] = backtesting_dict[f2]['Rev.'] + row["Rev. 2"]
    backtesting_dict[f2]['Ctrl'] = backtesting_dict[f2]['Ctrl'] + row["Ctrl 2"]
    
    backtesting_dict[f2]['Sig. str attempted'] = backtesting_dict[f2]['Sig. str attempted'] + row['Sig. str attempted2']
    backtesting_dict[f2]['Sig. str defended'] = backtesting_dict[f2]['Sig. str defended'] + row['Sig. str defended2']
    backtesting_dict[f2]['Sig. str absorbed'] = backtesting_dict[f2]['Sig. str absorbed'] + row['Sig. str absorbed2']
    
    backtesting_dict[f2]['Td attempted'] = backtesting_dict[f2]['Td attempted'] + row["Td attempted 2"]
    backtesting_dict[f2]['tdd'] = backtesting_dict[f2]['tdd'] + row['tdd2']
    
    backtesting_dict[f2]['time'] = backtesting_dict[f2]['time'] + row['time']
```

```python
def return_calculator(clf, threshold=0.5):
    investment = 100*len(final_df)
    abs_return = 0
    for i in oddsindex:
        odd1 = model_df2.loc[i]["fighter1_odds"]
        odd2 = model_df2.loc[i]["fighter2_odds"]
        actual_result = backtest_df.loc[i]["target"]
        fight = np.array(final_df.loc[i]).reshape(1, -1)
        predicted_proba = clf.predict_proba(fight)


        if threshold != 0.5:
            if predicted_proba[0][1] >= threshold:
                predicted_result = 1
            else:
                predicted_result = 0
            if predicted_result != actual_result:
                abs_return += 0
            else:

                if predicted_result == 0:

                    if int(odd2) < 0:
                        abs_return += 10000/abs(odd2)
                    else:
                        abs_return += odd2


                else:

                    if int(odd1) < 0:
                        abs_return += 10000/abs(odd1)
                    else:
                        abs_return += odd1

        else:
            predicted_result = clf.predict(fight)

            if predicted_result != actual_result:
                abs_return += 0
            else:
                #abs_return += 100
                if predicted_result == 0:

                    if int(odd2) < 0:
                        abs_return += 10000/abs(odd2)
                    else:
                        abs_return += odd2


                else:

                    if int(odd1) < 0:
                        abs_return += 10000/abs(odd1)
                    else:
                        abs_return += odd1

    return (abs_return/investment, clf, threshold)

        
```
