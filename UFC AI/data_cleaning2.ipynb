```python
#cleaning some data, will convert strings to ints, create a few more columns
#overall strikes fighter 1
sig_strike_landed1 = []
sig_strike_attempted1 = []
sig_strike_defended2 = []
for s in fights_data["Sig. str 1"]:
    s = s.split(" ")
    sig_strike_landed1.append(int(s[0]))
    sig_strike_attempted1.append(int(s[2]))
    delta = int(s[2]) - int(s[0])
    sig_strike_defended2.append(delta)
    
sig_strike_absorbed2 = sig_strike_landed1
```

```python
#same for fighter 2
sig_strike_landed2 = []
sig_strike_attempted2 = []
sig_strike_defended1 = []
for s in fights_data["Sig. str 2"]:
    s = s.split(" ")
    sig_strike_landed2.append(int(s[0]))
    sig_strike_attempted2.append(int(s[2]))
    delta = int(s[2]) - int(s[0])
    sig_strike_defended1.append(delta)
    
sig_strike_absorbed1 = sig_strike_landed1
```

```python
#specific targeted area strikes now, for fighter 1
head1 = []
body1 = []
leg1 = []
distance1 = []
clinch1 = []
ground1 = []

for i in range(0, len(fights_data)):
    head1.append(int((fights_data.loc[i, "Head 1"].split(" ")[0])))
    body1.append(int((fights_data.loc[i, "Body 1"].split(" ")[0])))
    leg1.append(int((fights_data.loc[i, "Leg 1"].split(" ")[0])))
    distance1.append(int((fights_data.loc[i, "Distance 1"].split(" ")[0])))
    clinch1.append(int((fights_data.loc[i, "Clinch 1"].split(" ")[0])))
    ground1.append(int((fights_data.loc[i, "Ground 1"].split(" ")[0])))

    
```

```python
#now the same for fighter 2
head2 = []
body2 = []
leg2 = []
distance2 = []
clinch2 = []
ground2 = []

for i in range(0, len(fights_data)):
    head2.append(int((fights_data.loc[i, "Head 2"].split(" ")[0])))
    body2.append(int((fights_data.loc[i, "Body 2"].split(" ")[0])))
    leg2.append(int((fights_data.loc[i, "Leg 2"].split(" ")[0])))
    distance2.append(int((fights_data.loc[i, "Distance 2"].split(" ")[0])))
    clinch2.append(int((fights_data.loc[i, "Clinch 2"].split(" ")[0])))
    ground2.append(int((fights_data.loc[i, "Ground 2"].split(" ")[0])))
```

```python
#now lets do grappling stats
td1 = []
td_attempted1 = []

td2 = []
td_attempted2 = []


for i in range(0, len(fights_data)):
    td1.append(int((fights_data.loc[i, "Td 1"].split(" ")[0])))
    td_attempted1.append(int((fights_data.loc[i, "Td 1"].split(" ")[2])))
    
    td2.append(int((fights_data.loc[i, "Td 2"].split(" ")[0])))
    td_attempted2.append(int((fights_data.loc[i, "Td 2"].split(" ")[2])))

    
```

```python
#now that we have made these changes, lets replace the relevant ones, and add the new ones for fighter 1
fights_data['Sig. str 1'] = sig_strike_landed1
fights_data['Sig. str attempted1'] = sig_strike_attempted1
fights_data['Sig. str absorbed1'] = sig_strike_absorbed1
fights_data['Sig. str defended1'] = sig_strike_defended1

fights_data['Head 1'] = head1
fights_data['Body 1'] = body1
fights_data['Leg 1'] = leg1
fights_data['Distance 1'] = distance1
fights_data['Clinch 1'] = clinch1
fights_data['Ground 1'] = ground1

fights_data['Td 1'] = td1
fights_data['Td attempted 1'] = td_attempted1
```

```python
#fighter 2
fights_data['Sig. str 2'] = sig_strike_landed2
fights_data['Sig. str attempted2'] = sig_strike_attempted2
fights_data['Sig. str absorbed2'] = sig_strike_absorbed2
fights_data['Sig. str defended2'] = sig_strike_defended2

fights_data['Head 2'] = head2
fights_data['Body 2'] = body2
fights_data['Leg 2'] = leg2
fights_data['Distance 2'] = distance2
fights_data['Clinch 2'] = clinch2
fights_data['Ground 2'] = ground2

fights_data['Td 2'] = td2
fights_data['Td attempted 2'] = td_attempted2
```

```python
#we forgot total strikes not just significant strikes
total_strike_landed1 = []
total_strike_landed2 = []

for s in fights_data["Total str. 1"]:
    s = s.split(" ")
    total_strike_landed1.append(int(s[0]))

for s in fights_data["Total str. 2"]:
    s = s.split(" ")
    total_strike_landed2.append(int(s[0]))
```

```python
fights_data["Total str. 1"] = total_strike_landed1
fights_data["Total str. 2"] = total_strike_landed2
```

```python
#lets check for missing data
fights_data.info()
#no null values
#but values of style "--" exist
#we will replace all of these with 0
fights_data2 = fights_data.replace(['--'], 0)
```

```python
#want to change control time to seconds for easier handling
import datetime

ctrl_times1 = []
ctrl_times2 = []

for t in fights_data2["Ctrl 1"]:
    try:
        ctr = int((datetime.datetime.strptime(t, "%M:%S") - datetime.datetime(1900, 1, 1)).total_seconds())
        ctrl_times1.append(ctr)
    except:
        ctrl_times1.append(t)
        
for t in fights_data2["Ctrl 2"]:
    try:
        ctr = int((datetime.datetime.strptime(t, "%M:%S") - datetime.datetime(1900, 1, 1)).total_seconds())
        ctrl_times2.append(ctr)
    except:
        ctrl_times2.append(t)
```

```python
#setting to dataframe
fights_data2["Ctrl 1"] = ctrl_times1
fights_data2["Ctrl 2"] = ctrl_times2
```

```python
#there was a slight error in scraping results
#the problem is that result 1 should be winners and 2 should be losers
#but result 2 is only Ls
#result 1 has about 1968 Ls
#probably something to do with website format change
#for every L in result1, we will replace result 2 of that corresponding row with a W
#the rest will be the same
#we will make a copy of the dataframe just in case
fights_data3 = fights_data2.copy()
```

```python
res1 = []
res2 = []
for r in fights_data3["result 1"]:
    if r == "L":
        res1.append(r)
        res2.append("W")
    elif r == "NC":
        res1.append(r)
        res2.append(r)
    elif r == "D":
        res1.append(r)
        res2.append(r)
    else:
        res1.append("W")
        res2.append("L")
```

```python
fights_data3["result 1"] = res1
fights_data3["result 2"] = res2
```

```python
fights_data3[fights_data3["result 1"] == "W"]
```

```python
#since draws and no contests are very rare, we will get rid of them entirely
fights_data4 = fights_data3.replace(["D", "NC"], float("NaN"))

fights_data4 = fights_data4.dropna()
```

```python
#as seen no more draws or unique values
print(set(fights_data4["result 1"]))
print(set(fights_data4["result 2"]))
```
