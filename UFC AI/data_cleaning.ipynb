```python
import pandas as pd
fighter_data = pd.read_pickle("fighter_data.pkl")
fights_data = pd.read_pickle("fights_data.pkl")

fights_data_cleaned = pd.concat([fights_data.drop(['stats'], axis=1), fights_data['stats'].apply(pd.Series)], axis=1).drop(['Fighter'], axis=1)
fights_data_cleaned


fights_data_cleaned


```

```python
result1 = []
fighter1 = []
sigstr1 = []
sigstrperc1 = []
head1 = []
body1 = []
leg1 = []
distance1 = []
clinch1 = []
ground1 = []
kd1 = []
totalstr1 = []
td1 = []
tdperc1 = []
subatt1 = []
rev1 = []
ctrl1 = []

time1 = []
method1 = []
date1 = []



result2 = []
fighter2 = []
sigstr2 = []
sigstrperc2 = []
head2 = []
body2 = []
leg2 = []
distance2 = []
clinch2 = []
ground2 = []
kd2 = []
totalstr2 = []
td2 = []
tdperc2 = []
subatt2 = []
rev2 = []
ctrl2 = []

time2 = []
method2 = []
date2 = []


for index, row in fights_data_cleaned.iterrows():
    if (index % 2) == 0:
        result1.append(row['result'])
        fighter1.append(row['fighter'])
        sigstr1.append(row['Sig. str'])
        sigstrperc1.append(row['Sig. str. %'])
        head1.append(row['Head'])
        body1.append(row['Body'])
        leg1.append(row['Leg'])
        distance1.append(row['Distance'])
        clinch1.append(row['Clinch'])
        ground1.append(row['Ground'])
        kd1.append(row['KD'])
        totalstr1.append(row['Total str.'])
        td1.append(row['Td'])
        tdperc1.append(row['Td %'])
        subatt1.append(row['Sub. att'])
        rev1.append(row['Rev.'])
        ctrl1.append(row['Ctrl'])
        
        time1.append(row['times'])
        method1.append(row['methods'])
        date1.append(row['date'])
        
    else:
        result2.append(row['result'])
        fighter2.append(row['fighter'])
        sigstr2.append(row['Sig. str'])
        sigstrperc2.append(row['Sig. str. %'])
        head2.append(row['Head'])
        body2.append(row['Body'])
        leg2.append(row['Leg'])
        distance2.append(row['Distance'])
        clinch2.append(row['Clinch'])
        ground2.append(row['Ground'])
        kd2.append(row['KD'])
        totalstr2.append(row['Total str.'])
        td2.append(row['Td'])
        tdperc2.append(row['Td %'])
        subatt2.append(row['Sub. att'])
        rev2.append(row['Rev.'])
        ctrl2.append(row['Ctrl'])
        
        time2.append(row['times'])
        method2.append(row['methods'])
        date2.append(row['date'])
```

```python
from pandas import DataFrame
rows = [result1, fighter1, sigstr1, sigstrperc1, head1, body1, leg1, distance1, 
        clinch1, ground1, kd1, totalstr1, td1, tdperc1, subatt1, rev1, ctrl1, time1, method1, date1, 
        result2, fighter2, sigstr2, sigstrperc2, head2, body2, leg2, distance2,
        clinch2, ground2, kd2, totalstr2, td2, tdperc2, subatt2, rev2, ctrl2, time2, method2, date2]
df = DataFrame (rows).transpose()
df.columns = ['result 1', 'fighter 1', 'Sig. str 1', 'Sig. str. % 1',
              'Head 1', 'Body 1', 'Leg 1', 'Distance 1', 'Clinch 1', 'Ground 1', 
             'KD 1', 'Total str. 1', 'Td 1', 'Td % 1', 'Sub. att 1', 'Rev. 1', 'Ctrl 1', 
              'time1', 'method1', 'date1',
             'result 2', 'fighter 2', 'Sig. str 2', 'Sig. str. % 2',
              'Head 2', 'Body 2', 'Leg 2', 'Distance 2', 'Clinch 2', 'Ground 2', 
             'KD 2', 'Total str. 2', 'Td 2', 'Td % 2', 'Sub. att 2', 'Rev. 2', 'Ctrl 2',
             'time2', 'method2', 'date2']
df
df.to_pickle('fights-data-cleaned.pkl')
```
