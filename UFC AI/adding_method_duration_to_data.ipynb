```python
import pandas as pd
method_duration_data = pd.read_pickle('method_duration_data.pkl')
method_duration_data['date']
```

```python
fights_strings_removed = pd.read_pickle('fights_strings_removed.pkl')
fights_strings_removed['date']
```

```python
method_duration_data['duration']
```

```python
method = []
duration = []
for index, row in fights_strings_removed.iterrows():
    data = method_duration_data.loc[((method_duration_data['fighter_1'] == row['fighter 1'])
                                     & (method_duration_data['fighter_2'] == row['fighter 2']))  
                                    | ((method_duration_data['fighter_1'] == row['fighter 2'])
                                       & (method_duration_data['fighter_2'] == row['fighter 1']))]
    
    method.append(data['method'].values)
    duration.append(data['duration'].values)
    
print(len(method))
print(len(duration))
```

```python
count = 0
for arr in method:
    if len(arr) == 0:
        count+=1
print(count)

count = 0
print(count)
for arr in duration:
    if len(arr) == 0:
        count+=1
print(count)
    
fights_strings_removed['method'] = method
fights_strings_removed['duration'] = duration
fights_strings_removed
```
