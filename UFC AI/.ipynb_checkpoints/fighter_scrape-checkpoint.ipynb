```python
import numpy as np
import pandas as pd
import urllib.request as urllib
from bs4 import BeautifulSoup
import time
import random
import re

fighter_info = pd.DataFrame()

full_name = []

height = []

weight = []

reach = []

stance = []

mma_record = []

belt = []

alphabet = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]

for i in alphabet:
    html = urllib.urlopen("http://ufcstats.com/statistics/fighters?char=" + i + "&page=all")
    soup = BeautifulSoup(html.read())
    sections = soup.find_all("tr")

    for section in sections:

        first_name = None
        last_name = None
        h = None
        wt = None
        r = None
        st = None
        wins = None
        losses = None
        draws = None

        

        subsections = section.find_all("td",{"class":"b-statistics__table-col"})
        if(len(subsections) == 11):
            if section.find("img") != None:
                belt.append(True)
            else:
                belt.append(False)
            first_name = subsections[0].get_text().strip()
            last_name = subsections[1].get_text().strip()
            h = subsections[3].get_text().strip()
            wt = subsections[4].get_text().strip()[:-5]
            r = subsections[5].get_text().strip()
            st = subsections[6].get_text().strip()
            wins = subsections[7].get_text().strip()
            losses = subsections[8].get_text().strip()
            draws = subsections[9].get_text().strip()

            full_name.append(first_name + " " + last_name)
            height.append(h)
            weight.append(wt)
            reach.append(r)
            stance.append(st)
            mma_record.append([wins, losses, draws])
        
```

```python
for h in height:
    if h != "--":
        h = (12 * int(h[0])) + (int(h[2:-1])/12)
```

```python
reach2 = []
for r in reach:
    reach2.append(r[:-1])
```

```python
height2 = []
for h in height:
    if h != "--":
        height2.append((12 * int(h[0])) + (int(h[2:-1])))
    else: 
        height2.append(h)
height2
```

```python
fighter_info = pd.DataFrame()
fighter_info["Name"] = full_name
fighter_info["Height"] = height2
fighter_info["Weight"] = weight
fighter_info["Reach"] = reach2
fighter_info["Stance"] = stance
fighter_info["Record"] = mma_record
fighter_info["Belt"] = belt

fighter_info
```

```python
#handling missing data
#first weight, using mean
fighter_info["Weight"] = pd.to_numeric(fighter_info["Weight"], errors='coerce')
fighter_info["Weight"] = fighter_info["Weight"].fillna((fighter_info["Weight"].mean()))
```

```python
#now height
fighter_info["Height"] = pd.to_numeric(fighter_info["Height"], errors='coerce')
fighter_info["Reach"] = pd.to_numeric(fighter_info["Reach"], errors='coerce')
```

```python
fighter_agg = fighter_info.groupby("Weight").agg({"Height" : "mean", "Reach" : "mean"})
```

```python
fighter_agg
```

```python
# we will work with height though
fighter_info["Height"] = fighter_info["Height"].fillna((fighter_info["Height"].mean()))
```

```python
fighter_info["Stance"].mode()
```

```python
fighter_info["Stance"] = fighter_info["Stance"].fillna((fighter_info["Stance"].mode()))
```

```python
for h in fighter_info["Stance"]:
    if h == "":
        h = fighter_info["Stance"].mode()
```

```python
stance2 = []
for s in fighter_info["Stance"]:
    if s == "":
        stance2.append("Orthodox")
    else:
        stance2.append(s)
stance2
```

```python
fighter_info["Stance"] = stance2
```

```python
fighter_df = pd.DataFrame()
fighter_df["Name"] = fighter_info["Name"]
fighter_df["Height"] = fighter_info["Height"]
fighter_df["Weight"] = fighter_info["Weight"]
fighter_df["Stance"] = fighter_info["Stance"]
fighter_df["Record"] = fighter_info["Record"]
fighter_df
```

```python
#pickling away for preprocessing
fighter_df.to_pickle("fighter_data.pkl")
```
