```python
import numpy as np
import pandas as pd
```

```python
features = pd.read_pickle("features.pkl")
target = pd.read_pickle("target.pkl")
```

```python
#lets split the data into train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.1, random_state=123)
```

```python
#lets normalize the features
from sklearn.preprocessing import MinMaxScaler

X_train = MinMaxScaler().fit_transform(X_train)
X_test = MinMaxScaler().fit_transform(X_test)
```

```python
#lets build an implementation of Logistic Regression from scratch using gradient descent
#sigmoid function
def sigmoid(data):
    return 1 / (1 + np.exp(-data))
```

```python
#gradient descent
import statsmodels.api as sm


def gradient_descent(x, y, alpha, iterations):
    
    #adding a column of zeros to deal with intercept, not just gradient
    x2 = sm.add_constant(x)
    observations = len(x2)
    
    #initializing all slopes to zero
    coeffs = np.zeros(x2.shape[1])
    
    #begin iterations
    for i in range(0, iterations):
        #current predictions
        ypred = np.matmul(x2, coeffs) #multiplies matrix of coefficients with matrix of data
        
        #applies sigmoid function on the predictions to get probabilities, and turn it from a 
        #linear regression to logistic regression
        ypred = sigmoid(ypred)
        
        #derivative of loss function (mean square error)
        der = -(np.matmul((y - ypred), x2))
        
        #update coefficients
        coeffs = coeffs - (alpha * der)
        
    return coeffs
```

```python
def predict_logistic(coeffs, features):
    #adding column of 1 to data again
    features = sm.add_constant(X_test)
    preds = sigmoid(np.matmul(features, coeffs))
    
    return preds
    
```

```python
#lets test out some arbitrary learning rate and iteration number to see if its working
coeffs = gradient_descent(X_train, y_train.values.reshape(5278), 0.001, 100000)
```

```python
#looks like its working pretty well
ypred = np.round(predict_logistic(coeffs, X_test))

from sklearn.metrics import accuracy_score

accuracy_score(y_test, ypred)
```

```python
#comparing to sklearn's built in Logistic Regression to see how well it is doing
from sklearn.linear_model import LogisticRegression

#change dthe penalty because by default sklearn includes some regularization, so would
#not be a fair comparison
logreg = LogisticRegression(penalty='none').fit(X_train, y_train)


ypred = logreg.predict(X_test)
accuracy_score(y_test, ypred)
```

```python
import pickle
# save the classifier
with open('test_classifier.pkl', 'wb') as fid:
    pickle.dump(logreg, fid)    
```

looks like our performs at a very similar level, so it must be correct. Let's try to look for a learning rate
or iteration number that may make it better

```python
lrs = [1, 0.1, 0.001, 0.0001]
iterations = [100, 1000, 10000, 100000, 10000000]

accuracies = []
alphas = []
iters = []

for i  in lrs:
    for j in iterations:
        coeffs = gradient_descent(X_train, y_train.values.reshape(5278), i, j)
        ypred = np.round(predict_logistic(coeffs, X_test))
        
        accuracies.append(accuracy_score(y_test, ypred))
        alphas.append(i)
        iters.append(j)
        

```

```python
df = pd.DataFrame()

df["learning rate"] = alphas
df["iterations"] = iters
df["accuracy"] = accuracies

df
```

```python
#lets include regularization in the model to see if it does better
logreg = LogisticRegression(penalty='l2').fit(X_train, y_train)


ypred = logreg.predict(X_test)
accuracy_score(y_test, ypred)
```

```python
#worse with ridge, lets try lasso
#had to use a different solver because the default solver did not include lasso

logreg = LogisticRegression(penalty='l1', solver = 'saga').fit(X_train, y_train)


ypred = logreg.predict(X_test)
accuracy_score(y_test, ypred)
```

```python
#now lets compare this model to other models
#lets try something simple like k nearest neighbors
from sklearn.neighbors import KNeighborsClassifier

#good practice to have neighbors at square root value of number of datapoints
knn = KNeighborsClassifier(n_neighbors= int(np.round(np.sqrt(X_train.shape[0])))).fit(X_train, y_train)
ypred = knn.predict(X_test)
accuracy_score(y_test, ypred)
```

```python
#not as good as logistic regression, thus algorithm is probably too simple for the task at hand
#will try decision tree
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier().fit(X_train, y_train)
ypred = tree.predict(X_test)
accuracy_score(y_test, ypred)
```

```python
#the basic tree is not that good, but it can probably be improved with some tuning
print(tree.get_depth())
#depth of this tree is 16, lets try searching for optimal tree depth, and prune
```

```python
accuracies = []
depth = [i for i in range(1, 17)]
for i in depth:
    tree = DecisionTreeClassifier(max_depth=i).fit(X_train, y_train)
    
    test_pred = tree.predict(X_test)

    acc = accuracy_score(y_test, test_pred)
    accuracies.append(acc)
    
df = pd.DataFrame()
df["depth"] = depth
df["testing error"] = accuracies
```

```python
df
```

```python
#looks like the best depth for the tree is at 7, I think an ensemble model will perform better
#lets try bagging wigth random forest
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier().fit(X_train, y_train)
ypred = rf.predict(X_test)
accuracy_score(y_test, ypred)
```

```python
#this is much better than the decision tree, may be worth tuning the model to improve it
from sklearn.model_selection import GridSearchCV

params = {"n_estimators": [100, 150, 300, 500], "max_depth":range(1, 17), "min_samples_split":range(2, 10),
         "min_samples_leaf":range(1,5), "min_impurity_decrease":[0, 1]}

grid = GridSearchCV(RandomForestClassifier(), param_grid=params)
grid.fit(X_train, y_train)
```
