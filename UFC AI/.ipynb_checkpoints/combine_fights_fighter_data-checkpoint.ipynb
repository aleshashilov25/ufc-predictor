```python
#fighter 1
heights1 = []
weights1 = []
stance1 = []


for f in fights_data["fighter 1"]:
    try:
        h = float(fighter_data[fighter_data["Name"] == f]["Height"])
        w = float(fighter_data[fighter_data["Name"] == f]["Weight"])
        s = str(fighter_data[fighter_data["Name"] == f]["Stance"]).split(" ")[4].split()[0]
    except:
        h = 0
        w = 0
        s = 0
    heights1.append(h)
    weights1.append(w)
    stance1.append(s)
```

```python
#fighter 2
heights2 = []
weights2 = []
stance2 = []

for f in fights_data["fighter 2"]:
    try:
        h = float(fighter_data[fighter_data["Name"] == f]["Height"])
        w = float(fighter_data[fighter_data["Name"] == f]["Weight"])
        s = str(fighter_data[fighter_data["Name"] == f]["Stance"]).split(" ")[4].split()[0]
    except:
        h = 0
        w = 0
        s = 0
    heights2.append(h)
    weights2.append(w)
    stance2.append(s)
```

```python
fights_data["Height 1"] = heights1
fights_data["Weight 1"] = weights1
fights_data["Stance 1"] = stance1

fights_data["Height 2"] = heights2
fights_data["Weight 2"] = weights2
fights_data["Stance 2"] = stance2
```

```python
#some missing data
#replace any zeros, for height we will take the mean
heights1 = []
for h in fights_data["Height 1"]:
    if h == 0:
        heights1.append(fights_data["Height 1"].mean())
    else:
        heights1.append(h)

heights2 = []
for h in fights_data["Height 2"]:
    if h == 0:
        heights2.append(fights_data["Height 2"].mean())
    else:
        heights2.append(h)

fights_data["Height 1"] = heights1
fights_data["Height 2"] = heights2
```

```python
#for stance most people are orthodox so we will put that in for missing stances
```

```python
stance1 = []
stance2 = []

for s in fights_data["Stance 1"]:
    if s == 0:
        stance1.append("Orthodox")
    else:
        stance1.append(s)

for s in fights_data["Stance 2"]:
    if s == 0:
        stance2.append("Orthodox")
    else:
        stance2.append(s)

fights_data["Stance 1"] = stance1
fights_data["Stance 2"] = stance2
```

```python
#now weight is a lil trickier, fights are in a weight class, 
#so if weight 1 is missing, we will put the corresponding weight 2 value
weight1 = []
weight2 = []
```

```python
for i in fights_data.index:
    try:
        if fights_data["Weight 1"][i] == 0:
            weight1.append(fights_data["Weight 2"][i])
        else:
            weight1.append(fights_data["Weight 1"][i])
    except:
        pass
        
for i in fights_data.index:
    try:
        if fights_data["Weight 2"][i] == 0:
            weight2.append(fights_data["Weight 1"][i])
        else:
            weight2.append(fights_data["Weight 2"][i])
    except:
        pass
        
fights_data["Weight 1"] = weight1
fights_data["Weight 2"] = weight2
    
```

```python
#records
#fighter 1
records1 = []


for f in fights_data["fighter 1"]:
    try:
        r = list(fighter_data[fighter_data["Name"] == f]["Record"])[0]
    except:
        r = 0
    records1.append(r)
```

```python
records2 = []


for f in fights_data["fighter 2"]:
    try:
        r = list(fighter_data[fighter_data["Name"] == f]["Record"])[0]
    except:
        r = 0
    records2.append(r)
```

```python
fights_data["Record 1"] = records1
fights_data["Record 2"] = records2
```
